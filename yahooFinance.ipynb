```python
#Display in dataframe using YahooFinancialsb

import pandas
from yahoofinancials import YahooFinancials
import datetime
```

```python
data3 = pd.DataFrame()
end_date = (dt.datetime.today()).strftime("%Y-%m-%d")
beg_date = (dt.datetime.today() - dt.timedelta(1825)).strftime("%Y-%m-%d")     #1825 = 5 years
tickers = ["AAPL", "MSFT", "CSCO"]
for ticker in tickers:
  yf2 = YahooFinancials(ticker)
  json_obj = yf2.get_historical_price_data(beg_date, end_date, "daily")
  ohsdata3 = json_obj[ticker]["prices"]
  temp = pd.DataFrame(ohsdata3)[["formatted_date", "adjclose"]]
  temp.set_index("formatted_date", inplace=True)
  data3[ticker] = temp["adjclose"]
```

```python
!pip install alpha_vantage  --quiet

from alpha_vantage.timeseries import TimeSeries
import pandas as pd
import time      #Alpha Vantage only allow 5 API call at 1 time to bypass it use time
```

```python
key_path = "/content/drive/MyDrive/Colab Notebooks/Scp_ud/AVAPI.txt"
```

```python
#Extract data for single tracker
ts = TimeSeries(key=open(key_path, "r").read(), output_format="pandas")
data4 = ts.get_daily(symbol="MSFT", outputsize="full")[0]                    #compact returns only the latest 100 data points in the intraday time series; full returns the full-length intraday time series. The "compact" option is recommended if you would like to reduce the data size of each API call.
data4.columns = ["open", "high", "low", "close", "volume"]
data4 = data4.iloc[::-1]    #to display the date in descending order
data4
```

```python
# Extract historical closing price for multiple tricker

all_ticker = ["MSFT", "AAPL", "CSCO", "AMZN", "GOOG", "FB"]
close_prices = pd.DataFrame()
api_call_count = 1
starttime = time.time()

for ticker in all_ticker: 
  ts = TimeSeries(key=open(key_path, "r").read(), output_format="pandas")
  data = ts.get_intraday(symbol=ticker, interval="60min", outputsize="compact")[0]
  api_call_count +=1
  data.columns = ["open", "high", "low", "close", "volume"]
  data = data[::-1]
  close_prices[ticker] = data["close"]
  if api_call_count == 5:
    api_call_count = 1
    time.sleep(60 - ((time.time()- starttime)%60.0))   # tells you how much time is remaining before the start of the new cycle. For example if your loop execution takes 6 seconds, this command will ensure that your code sleeps for the next 54 seconds before firing up.
```
